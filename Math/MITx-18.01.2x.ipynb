### Unit 4, Integration Techniques

Integrals of trig powers

Evaluating $\int sin^{m}(\theta)cos^{n}(\theta)d\theta$ (shows up in Fourier series)


**The easy case is when the exponents are odd:**

Take n = 1; $\int sin^{m}(x)cos(x)dx$, do substitution, make $ u = sin(x)$ , $du = cos(x)$.
Then the integral becomes $\int u^{m}du$

Take the antiderivative of that for $ \int \dfrac{u^{m+1}}{m+1}du$. Plug back in sin and add + C.

When m and n > 1, but m or n is odd. use $cos^2(\theta) + sin^2(\theta) = 1 $ to get rid of the largest even power.

ex/ $\int sin^{3}(x) dx $ = $\int (1-cos^{2})sin(x)dx$ . From there do substitution and all that jazz.


**The harder case is when the exponents are even:**'

You will need to use half angle formulas

$ sin^2(a) = \dfrac{1-cos(2a)}{2} $

$ cos^2(a) = \dfrac{1+cos(2a)}{2} $

ex/ $\int cos^2(x) dx $ = $\int \dfrac{1+cos(xa)}{2}$  = $ \int \dfrac{1}{2} + \dfrac{cos(2x)}{2}$

Take the antiderivative of each and you have

$  \dfrac{x}{2} + \dfrac{sin(2x)}{4} +C$ with the sin being over 4 because you need to do the chain rule with sin(2x)

Partial Fractions, Aka Partial Fraction Decomposition

Dealing with rational functions ( A rational function being ratios of polynomials: $\dfrac{P(x)}{Q(x)}$ where both P and Q are polynomials), by splitting them into easier pieces. Most of the manipulation is algebraic. Make sure they are polynomials, a sqrt in the denominator over a polynomial expression is not allowed for this method.

If told to integrate a function like $ \dfrac{4x-1}{x^{2}+x-2} $.

The way to get this into an easy form is using the cover up method. the degree of the polynomial in the numerator needs to be less than the denominator degree, and the denominator degree needs to be able to factor completely. 

First: factor the denominator

Second: set up equations where there's an A and B over the factored denominator, like this $ \dfrac{4x-1}{(x-1)(x+2)} =  \dfrac{A}{x-1} +  \dfrac{B}{x+2} $

Third: solve for A and B. I'll solve for A here but it works likewise for B.

Start isolating A by multiplying the equation by (x-1) to get $ \dfrac{4x-1}{(x+2)} =  {A} +  (\dfrac{B}{x+2}*(x-1)) $

Then plug in an x that it would take to zero B (or A if doing B side) which in this case is 1. (the limit as x approaches 1, the reason why it's a limit is because putting in x = 1 into the original equation makes the denominator zero.)

Then what you have is $ \dfrac{4(1) - 1}{(1)+2} = A $, which means A = 1.

Do the same process for B and you get $\dfrac{4x-1}{x^{2}+x-2} = \dfrac{1}{x-1} +  \dfrac{3}{x+2}$

From there the integration is easy.

---------------------------

If there are repeated linear factors in the denominator $Q(x) = (x-a)^{n}$ : then you set it up as

$\dfrac{P(x)}{Q(x)} = \dfrac{A_{1}}{(x-a)} + \dfrac{A_{2}}{(x-a)^{2}} + ... +  \dfrac{A_{n}}{(x-a)^{n}} $

The whole point is to solve for unknowns, if you ever get to where you have one unknown left you can pick an x not already used by another variable, plug it in and find the last unknown.

____

If there are quadratic factors,

$\dfrac{x^{2}}{(x-1)(x^{2}+1)} $

The setup becomes

$\dfrac{A}{(x-1)} + \dfrac{B_{x}+C}{(x^{2}+1)} $ where Bx is the quadratic part and C is the constant term.

Use cover up for A, then clear the denominator, then solve for B and C looking at the coefficients of x^2 and x^0.
The professor in the videos said to look at the two ends and not the intermediates like x^1 because the ends contain the most information.

Integration By Parts

This is basically a rearrangement of the product rule for derivatives except its now for integration, and it can be useful for finding otherwise seemingly impossible antiderivatives.

$\int u v' dx  = uv - \int u'vdx$

For definite integrals its

$\int^{a}_{b} u v' dx  = uv|^{a}_{b} - \int^{a}_{b} u'vdx$

----

Example with log function:

Given $\int ln(x) dx$ set $v'$ = 1 so that it's antiderivative $ v = x $ and it makes the left hand side easy to evaluate.

Set $ln(x) = u$ so $u' = \dfrac{1}{x}$

Then you have $ \int ln(x) 1 dx = xln(x) - \int x \dfrac{1}{x}dx$

$x * \dfrac{1}{x} = 1$, so in the end you get $xln(x) - x+C$ as the answer to the equation

A large part of integration by parts is picking a u that is easy to integrate and a v' that is easy to take the antiderivative of.

Arc Lengths

To find the length of a curved line, you take the pythagorean theorem where a and b in$\sqrt(a^{2}+b^{2})$ is $\Delta x $ and $ \Delta y$.

As the differences become infinitesimal, and then summed over, the sum will give the exact length of the curved line.

The new version of the pythagorean theorem is called the differential arc length, denoted ds

$ds = \sqrt(dx^{2}+dy^{2})$

can also be written

$ds = \sqrt(1+(\dfrac{dy}{dx})^{2}) dx$ the dx is factored out

$ds = \sqrt(1+(f'(x))^{2}) dx$

$ds = \sqrt((\dfrac{dx}{dy})^{2}+1) dy$ dy is factored out

$ds = \sqrt(((f^{-1})'(y))^{2}+1) dy$

The arc lengths, of a curve y=f(x) is

$\int^{sn}_{s0}ds$, where $s_{n}$ is the end of the line and $s_{0}$ is the start
